# PairGrid

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris['species'].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
sns.pairplot(iris, hue='species')
```

```python
tips.head()
```

```python
h = sns.FacetGrid(tips, col='time', row='smoker')
h.map(plt.hist, 'total_bill')
```

```python
h = sns.FacetGrid(tips, col="time",  row="smoker",hue='sex')
# Observe como os argumentos vm aps a chamada do plt.scatter
h = h.map(plt.scatter, "total_bill", "tip").add_legend()
```
